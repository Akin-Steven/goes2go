# Demo: Image Brighten/Darken with Gamma Correction

```python
from goes2go.data import goes_latest, goes_nearesttime
from goes2go.rgb import normalize, TrueColor, NaturalColor
import matplotlib.pyplot as plt
```

```python
# Obtain a GOES ABI multichannel image
g = goes_nearesttime("2020-12-13 18:00", satellite=16, product="ABI", domain="F")
```

---
Display Natural Color Images with different Gamma values

```python
gammas = [0.5, 0.8, 1, 1.2, 1.5]

fig, axes = plt.subplots(1, len(gammas), figsize=[15, 3.5], dpi=100)
for gamma, ax in zip(gammas, axes):
    nc = NaturalColor(g.thin(5), gamma)
    ax.imshow(nc.NaturalColor)
    ax.axis("off")
    ax.set_title(f"gamma={gamma:.1f}")
    print("finished gamma=", gamma)
plt.subplots_adjust(wspace=0.01)
plt.suptitle("Pseudo Green (pseudoGreen=True)", fontweight="bold")

plt.savefig(f"../docs/_static/gamma_demo_NaturalColor-PsuedoGreen", bbox_inches="tight")
```

```python
gammas = [0.5, 0.8, 1, 1.2, 1.5]

fig, axes = plt.subplots(1, len(gammas), figsize=[15, 3.5], dpi=100)
for gamma, ax in zip(gammas, axes):
    nc = NaturalColor(g.thin(5), gamma, pseudoGreen=False)
    ax.imshow(nc.NaturalColor)
    ax.axis("off")
    ax.set_title(f"gamma={gamma:.1f}")
    print("finished gamma=", gamma)
plt.subplots_adjust(wspace=0.01)
plt.suptitle('"Veggie" Green (pseudoGreen=False)', fontweight="bold")

plt.savefig(f"../docs/_static/gamma_demo_NaturalColor-VeggieGreen", bbox_inches="tight")
```
