# GOES ABI field of view for domains. 
[Issue #4](https://github.com/blaylockbk/goes2go/issues/4)

```python
from goes2go.data import goes_latest
from goes2go.tools import abi_crs, field_of_view
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
from toolbox.cartopy_tools import common_features
```

```python
G = goes_latest(satellite=17)
```

```python
FOV_full, FOV_dom, crs = field_of_view(G)
```

```python
ax = common_features(
    OCEAN=True,
    LAND=True,
    crs=ccrs.PlateCarree(central_longitude=-100),
    figsize=[10, 5],
    dark=True,
)

ax.add_geometries(
    [FOV_full],
    crs=crs,
    facecolor="tab:purple",
    edgecolor="tab:purple",
    alpha=0.3,
    linewidth=3,
)
ax.add_geometries(
    [FOV_dom],
    crs=crs,
    facecolor="tab:green",
    edgecolor="tab:green",
    alpha=0.3,
    linewidth=3,
)
```

```python
ax = common_features(
    OCEAN=True,
    LAND=True,
    crs=ccrs.PlateCarree(central_longitude=-100),
    figsize=[10, 5],
    dark=True,
)

G = goes_latest(satellite=16, domain="M1")

FOV_full, FOV_dom, crs = field_of_view(G)

G2 = goes_latest(satellite=16, domain="M2")

_, FOV_dom2, _ = field_of_view(G2)

ax.add_geometries(
    [FOV_full],
    crs=crs,
    facecolor="tab:purple",
    edgecolor="tab:purple",
    alpha=0.3,
    linewidth=3,
)
ax.add_geometries(
    [FOV_dom, FOV_dom2],
    crs=crs,
    facecolor="tab:green",
    edgecolor="tab:green",
    alpha=0.3,
    linewidth=3,
)
```
